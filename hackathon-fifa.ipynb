```python
# Importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the Dataset
df=pd.read_csv("football_data.csv")
```

```python
#exploring the data
df.columns
```

```python
# Dropping some columns
df.drop(['Unnamed: 0','Photo','Flag','Club Logo','Loaned From'],axis=1,inplace=True)
df.head()
```

```python
df.head()
```

```python
# Quick statistical analysis of Dataset
df.describe()
```

```python
# Checking for null values
df.info()
```

```python
# Number of numeric columns
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numeric_df = df.select_dtypes(include=numerics)
print(numeric_df)
len(numeric_df.columns)
```

```python
# Manipulation of Height column
df['Height'].replace(to_replace="'",value=".",inplace=True,regex=True)
```

```python
df['Height'].fillna(df['Height'].median(), inplace=True)
df['Height'].isnull().sum()
```

```python
df['Height']=df.Height.apply(lambda x:float(x))
df.Height.head()
```

```python
df['Height'].dtypes
```

```python
sns.set()
df['Height'].hist(bins = 10)
```

```python
# To show different height of players participating in the FIFA
plt.figure(figsize = (13, 8))
sns.countplot(x = 'Height', data = df, palette = 'dark')
plt.title(label = 'Count of players on Basis of Height')
plt.xlabel(xlabel = 'Height in Foot per inch')
plt.ylabel(ylabel = 'Count')
plt.show()
```

From above plot we can observe that maximum number of players are of height 6.0

```python
# Manipulation of weight column
df.Weight.replace(to_replace ="lbs",value ="",regex =True,inplace =True)
df.Weight.head()
```

```python
df.Weight.fillna(df.Weight.median(),inplace =True)
df.Weight.isnull().sum()
```

```python
#Converting the weights to Kg's
df.Weight =df.Weight.apply(lambda x :float(x)/2.20462)
df.Weight.head()
```

```python
sns.set()
df['Weight'].hist(bins = 10)
```

```python
# To show Different body weight of the players participating in the FIFA

plt.figure(figsize = (30,8))
sns.countplot(x = 'Weight', data = df)
plt.title('Different Weights of the Players Participating in FIFA')
plt.xlabel('Heights associated with the players')
plt.ylabel('count of Players')
plt.show()
```

```python
# Manipulation of Value column
df['Value'] = df['Value'].apply( lambda x : float(x.replace('€','').replace('K','')) * 1000 if 'K' in x else float(x.replace('€','').replace('M','')) * 1000000 if 'M' in x else x.replace('€',''))
df['Value'] = df['Value'].astype(float)
```

```python
df.Value =df.Value.apply(lambda x :float(x)*1e6)
df.Value.head()
```

```python
sns.set()
df['Value'].hist(bins = 10)
```

```python
## Comparing the players’ Values

plt.rcParams['figure.figsize'] = (15, 5)
sns.distplot(df['Value'], color = 'blue')
plt.xlabel('Value Range for Players')
plt.ylabel('Count of the Players')
plt.title('Distribution of Values of Players')
plt.show()
```

```python
#Manipulation of Wage column
df['Wage'] = df['Wage'].apply( lambda x : float(x.replace('€','').replace('K','')) * 1000 if 'K' in x else float(x.replace('€','')))
```

```python
df.Wage =df.Wage.apply(lambda x :float(x)*1e3)
df.Wage.head()
```

```python
sns.set()
df['Wage'].hist(bins = 10)
```

```python
## Comparing the players’ Wages

plt.rcParams['figure.figsize'] = (15, 5)
sns.distplot(df['Wage'], color = 'blue')
plt.xlabel('Wage Range for Players')
plt.ylabel('Count of the Players')
plt.title('Distribution of Wages of Players')
plt.show()
```

```python
# Correlation between some features using Heatmap
numerical_columns = df[['Height', 'Weight', 'Value', 'Age', 'Wage']]
sns.heatmap(numerical_columns.corr(), annot=True)
plt.show()
```

From the above plot we can see that Value and Wage are highly correlated.
